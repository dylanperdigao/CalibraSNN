```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')
```

```python
folders = [
    ('SM_1H_K2',11,"CSNN"),
    #'SM_1H_K3',
    #'SM_1H_K4',
    #'SM_1H_K5',
    ('MM_1H_K2',15,"CSNN"),
    #'MM_1H_K3',
    #'MM_1H_K4',
    #'MM_1H_K5',
    ('LM_1H_K2',23,"CSNN"),
    #'LM_1H_K3',
    #'LM_1H_K4',
    #'LM_1H_K5',
    #'SM_2H_K2',
    ('SM_2H_K3',16,"CSNN"),
    #'SM_2H_K4',
    ('MM_2H_K2',9,"CSNN"),
    #'MM_2H_K3',
    #'MM_2H_K4',
    #'LM_2H_K2',
    ('LM_2H_K3',23,"CSNN"),
    #'LM_2H_K4',
    ('SM_3H_K2',21,"CSNN"),
    ('MM_3H_K2',22,"CSNN"),
    ('LM_3H_K2',10,"CSNN"),
    # FFSNN
    ("FFSNN_1H",22,"FFSNN"),
    ("FFSNN_2H",25,"FFSNN"),
    ("FFSNN_3H",21,"FFSNN"),
    # CNN
    ("CNN_1H",1,"CNN"),
    ("CNN_2H",1,"CNN"),
    ("CNN_3H",1,"CNN"),
    # LFNN
    ("FFNN_1H",1,"FFNN"),
    ("FFNN_2H",1,"FFNN"),
    ("FFNN_3H",1,"FFNN"),
]

dfs = []
for f, steps,type in folders:
    results_df = pd.read_csv(f'../results/{f}.csv')
    model = f.split('-')[0]
    results_df['model'] = model
    results_df['steps'] = steps
    results_df['type'] = type
    dfs.append(results_df)
    print(f'Loaded {f} – {len(results_df)} results')
len(dfs)
```

```python
df = pd.concat(dfs)
print(df.shape)
df
```

```python
df_metrics = df.copy()
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
df_metrics[metrics[0]] = df_metrics['accuracy']
df_metrics[metrics[1]] = df_metrics['precision']
df_metrics[metrics[2]] = df_metrics['recall']
df_metrics[metrics[3]] = df_metrics['f1_score']
df_metrics = df_metrics.melt(id_vars=['type', 'dataset'], value_vars=metrics)

sns.set_theme(style="darkgrid")
sns.set_context("paper", font_scale=2)
fig = plt.figure(figsize=(16,8))
ax = plt.gca()
sns.violinplot(x="variable", y="value", hue="type", data=df_metrics, palette="tab10", cut=0, density_norm="width", ax=ax)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title="Model", loc="upper right", ncol=1)
ax.set_xlabel("Metric")
ax.set_ylabel("Value")
plt.tight_layout()
plt.show()
```

```python
df_fairness = df.copy()
metrics = ['$PE_{age}$', '$PE_{income}$', '$PE_{employment}$']
df_fairness[metrics[0]] = df_fairness['fpr_ratio_age']
df_fairness[metrics[1]] = df_fairness['fpr_ratio_income']
df_fairness[metrics[2]] = df_fairness['fpr_ratio_employment']
df_fairness = df_fairness.melt(id_vars=['type', 'dataset'], value_vars=metrics)
df_fairness = df_fairness[df_fairness['value'] > 0]


sns.set_theme(style="darkgrid")
sns.set_context("paper", font_scale=1.75)
fig = plt.figure(figsize=(16,8))
ax = plt.gca()
sns.violinplot(x="variable", y="value", hue="type", data=df_fairness, palette="tab10", cut=0, density_norm="width", ax=ax)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title="Model", loc="lower right", ncol=1)
ax.set_xlabel("Metric")
ax.set_ylabel("Value")
plt.tight_layout()
plt.show()
```

```python
df_energy = df.copy()
df_energy['joules'] = df_energy['energy_test']
df_energy['kilojoules'] = df_energy['energy_test'] / 1000
df_energy['watts_hours'] = df_energy['joules'] / 3600
df_energy['watts'] = df_energy['joules'] / df_energy['time_test']
df_energy['joules/steps'] = df_energy['joules'] / df_energy['steps']
df_energy['watts/steps'] = df_energy['watts'] / df_energy['steps']
metrics = ['Joules $[J]$', 'Watts $[W]$', 'Joules per Step', 'Watts per Step']
df_energy[metrics[0]] = df_energy['joules']
df_energy[metrics[1]] = df_energy['watts']
df_energy[metrics[2]] = df_energy['joules/steps']
df_energy[metrics[3]] = df_energy['watts/steps']
df_energy = df_energy.melt(id_vars=['type', 'dataset'], value_vars=metrics)

sns.set_theme(style="darkgrid")
sns.set_context("paper", font_scale=1.75)
fig = plt.figure(figsize=(16,8))
ax = plt.gca()
sns.violinplot(x="variable", y="value", hue="type", data=df_energy, palette="tab10", cut=0, density_norm="width", ax=ax)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title="Model", loc="upper right", ncol=1)
ax.set_xlabel("Metric")
ax.set_ylabel("Value")
ax.set_yscale('log')
plt.tight_layout()
plt.show()
```
