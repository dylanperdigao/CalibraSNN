```python
import pandas as pd
import sys

sys.path.append('..')
pd.set_option('display.max_columns', 100)
```

```python
folders = [
    ('SM_1H_K2',11),
    #'SM_1H_K3',
    #'SM_1H_K4',
    #'SM_1H_K5',
    ('MM_1H_K2',15),
    #'MM_1H_K3',
    #'MM_1H_K4',
    #'MM_1H_K5',
    ('LM_1H_K2',23),
    #'LM_1H_K3',
    #'LM_1H_K4',
    #'LM_1H_K5',
    # 2h
    #'SM_2H_K2',
    ('SM_2H_K3',16),
    #'SM_2H_K4',
    ('MM_2H_K2',9),
    #'MM_2H_K3',
    #'MM_2H_K4',
    #'LM_2H_K2',
    ('LM_2H_K3',23),
    #'LM_2H_K4',
    # 3h
    ('SM_3H_K2',21),
    ('MM_3H_K2',22),
    ('LM_3H_K2',10),
    # FFSNN
    ("FFSNN_1H",22),
    ("FFSNN_2H",25),
    ("FFSNN_3H",21),
    # CNN
    ("CNN_1H",1),
    ("CNN_2H",1),
    ("CNN_3H",1),
    # LFNN
    ("FFNN_1H",1),
    ("FFNN_2H",1),
    ("FFNN_3H",1),
]

dfs = []
for f, steps in folders:
    results_df = pd.read_csv(f'../results/{f}.csv')
    model = f.split('-')[0]
    results_df['model'] = model
    results_df['steps'] = steps
    dfs.append(results_df)
    print(f'Loaded {f} – {len(results_df)} results')
len(dfs)
```

```python
df = pd.concat(dfs)
print(df.shape)
df
```

```python
df_energy = df.copy()

df_energy['joules_train'] = df_energy['energy_train']
df_energy['watts_hours_train'] = df_energy['joules_train'] / 3600

df_energy['joules_test'] = df_energy['energy_test']
df_energy['watts_hours_test'] = df_energy['joules_test'] / 3600

df_energy['watts_train'] = df_energy['joules_train'] / df_energy['time_train']
df_energy['ratio_J_train'] = df_energy['joules_train'] / df_energy['steps']
df_energy['ratio_W_train'] = df_energy['watts_train'] / df_energy['steps']

df_energy['watts_test'] = df_energy['joules_test'] / df_energy['time_test']
df_energy['ratio_J_test'] = df_energy['joules_test'] / df_energy['steps']
df_energy['ratio_W_test'] = df_energy['watts_test'] / df_energy['steps']
df_energy = df_energy.groupby(['model']).agg({
    'time_train': ['mean', 'std'],
    'joules_train': ['mean', 'std'],
    'watts_train': ['mean', 'std'],
    'ratio_J_train': ['mean', 'std'],
    'ratio_W_train': ['mean', 'std'],
    'time_test': ['mean', 'std'],
    'joules_test': ['mean', 'std'],
    'watts_test': ['mean', 'std'],
    'ratio_J_test': ['mean', 'std'],
    'ratio_W_test': ['mean', 'std'],
})

metrics_names = ['time_test', 'joules_test', 'watts_test', 'ratio_J_test', 'ratio_W_test']
df_formatted = pd.DataFrame(index=df_energy.index)
for metric in metrics_names:
    df_formatted[metric] = (
        (df_energy[(metric, 'mean')]).round(2).astype(str) +
        ' $\\pm$ ' +
        (df_energy[(metric, 'std')]).round(2).astype(str)
    )
df_formatted.round(3)
```

```python
df_performance = df.copy()
df_performance = df_performance.groupby(['model']).agg({
    'accuracy': ['mean', 'std'],
    'f1_score': ['mean', 'std'],
    'precision': ['mean', 'std'],
    'recall': ['mean', 'std'],
    'fpr': ['mean', 'std'],
    'auc': ['mean', 'std'],
    'recall@5FPR': ['mean', 'std'],
    'fpr_ratio_age': ['mean', 'std'],
    'fpr_ratio_income': ['mean', 'std'],
    'fpr_ratio_employment': ['mean', 'std'],
})

df_formatted = pd.DataFrame(index=df_performance.index)

metrics = ['accuracy', 'precision', 'recall', 'fpr',  'f1_score', 'auc']
metrics_constraint = ['fpr_ratio_age', 'fpr_ratio_income', 'fpr_ratio_employment']

for metric in metrics+metrics_constraint:
    df_formatted[metric] = (
        (df_performance[(metric, 'mean')]*100).round(2).astype(str) +
        ' $\\pm$ ' +
        (df_performance[(metric, 'std')]*100).round(2).astype(str)
    )

df_formatted[metrics].to_latex('../tables/metrics_performance.tex', index=True, escape=False)
df_formatted[metrics_constraint].to_latex('../tables/metrics_fairness.tex', index=True, escape=False)

df_formatted.round(2)
```

```python
df_energy_train = df.copy().drop(columns=['energy_test', 'time_test'])
df_energy_train['type'] = 'train'
df_energy_train.rename(columns={'energy_train': 'energy', 'time_train': 'time'}, inplace=True)
df_energy_train['power'] = df_energy_train['energy'] / df_energy_train['time']
df_energy_train['ratio_energy'] = df_energy_train['energy'] / df_energy_train['steps']
df_energy_train['ratio_power'] = df_energy_train['power'] / df_energy_train['steps']

df_energy_test = df.copy().drop(columns=['energy_train', 'time_train'])
df_energy_test['type'] = 'test'
df_energy_test.rename(columns={'energy_test': 'energy', 'time_test': 'time'}, inplace=True)
df_energy_test['power'] = df_energy_test['energy'] / df_energy_test['time']
df_energy_test['ratio_energy'] = df_energy_test['energy'] / df_energy_test['steps']
df_energy_test['ratio_power'] = df_energy_test['power'] / df_energy_test['steps']

df_energy = pd.concat([df_energy_train, df_energy_test])
df_energy = df_energy.groupby(['type', 'model']).agg({
    'time': ['mean', 'std'],
    'energy': ['mean', 'std'],
    'power': ['mean', 'std'],
    'ratio_energy': ['mean', 'std'],
    'ratio_power': ['mean', 'std'],
})

df_formatted = pd.DataFrame(index=df_energy.index)
metrics_energy = ['time', 'energy', 'power', 'ratio_energy', 'ratio_power']
for metric in metrics_energy:
    df_formatted[metric] = (
        (df_energy[(metric, 'mean')]).round(2).astype(str) +
        ' $\\pm$ ' +
        (df_energy[(metric, 'std')]).round(2).astype(str)
    )

df_formatted[metrics_energy].to_latex('../tables/metrics_energy.tex', index=True, escape=False)
df_formatted.round(2)
```

```python
list_performance_metric_names = [('performance', 'accuracy'), ('performance', 'precision'), ('performance', 'recall'), ('performance', 'f1_score')]
list_fairness_metric_names = [('fairness', 'fpr_ratio_age'), ('fairness', 'fpr_ratio_income'), ('fairness', 'fpr_ratio_employment')]
list_metric_names = list_performance_metric_names + list_fairness_metric_names
for metric_type, metric_name in list_metric_names:
    df_metric = df.copy()
    df_metric = df_metric.groupby(['model', 'dataset']).agg({
        metric_name: ['mean', 'std'],
    })
    df_formatted = pd.DataFrame(index=df_metric.index)
    df_formatted[metric_name] = (
        (df_metric[(metric_name, 'mean')]*100).round(2).astype(str) +
        ' $\\pm$ ' +
        (df_metric[(metric_name, 'std')]*100).round(2).astype(str)
    )
    df_formatted = df_formatted.unstack(level='dataset')
df_formatted
```
